```python
import wave
import numpy as np
import matplotlib.pyplot as plt 

freq = [1.01, 1.82, 2.86, 3.70]
wave_len = [7.314, 3.657, 2.438, 1.829]

wave_len_np = np.array(wave_len)

plt.figure(1)
plt.plot(wave_len_np, freq, 'o', color= 'red', markersize=4)
plt.xlabel('Wavelength(m)', fontsize=14)
plt.ylabel('Frequency(Hz)', fontsize=14)
plt.title('Frequency vs Wavelength For Spring With L = 3.657m')
plt.show()

```

```python
plt.plot(1/wave_len_np, freq, 'o', color= 'red', markersize=4)
plt.xlabel('Inverse of Wavelength(1/m)', fontsize=14)
plt.ylabel('Frequency(Hz)', fontsize=14)
plt.title('Frequency vs Inverse of Wavelength For Spring With L = 3.657m')
plt.show()
```

```python
from scipy.optimize import curve_fit
import math

length = [68, 64.5, 60, 52, 44, 36, 28]
fund_freq = [143, 149, 159, 184, 219, 267, 345]
plt.plot(length, fund_freq, 'o', color= 'red', markersize=4, label='data')
plt.xlabel('Length(cm)', fontsize=14)
plt.ylabel('Fundmental Frequency(Hz)', fontsize=14)
plt.title('Fund. Frequency vs Length of Vibrating Wire For Constant T = 49.5N ')


def function_fit(length, a):
    frequency = a/length
    return frequency 

popt, pcov = curve_fit(function_fit, length, fund_freq)

a = popt[0]

print('The best fit parameters are:\n')
print(f'a = {a}')

xvals = np.linspace(25, 70, 200)
plt.plot(xvals, function_fit(xvals, a), color='black', linewidth=1.2, label='best-fit model')
plt.legend()
plt.show()
```

```python
tension = [69.1, 59.3, 49.5, 39.7, 29.9, 20.1]
fund_freq = [187, 174, 159, 143, 125, 100]
plt.plot(tension, fund_freq, 'o', color= 'red', markersize=4, label='data')
plt.xlabel('Tension(N)', fontsize=14)
plt.ylabel('Fundamental Frequency(Hz)', fontsize=14)
plt.title('Fund. Frequency vs Length of Vibrating Wire For Constant T = 49.5N ')


def function_fit(tension, a):
    frequency = a*(tension**(1/2))
    return frequency 

popt, pcov = curve_fit(function_fit, tension, fund_freq)

a = popt[0]

print('The best fit parameters are:\n')
print(f'a = {a}')

xvals = np.linspace(20, 70, 200)
plt.plot(xvals, function_fit(xvals, a), color='black', linewidth=1.2, label='best-fit model')
plt.legend()
plt.show()
```
