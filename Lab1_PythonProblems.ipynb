```python
from statistics import mean, stdev
import numpy as np
import matplotlib.pyplot as plt 
from scipy.optimize import curve_fit


voltage = [4.821, 5.493, 5.626, 4.951, 5.248, 5.588, 4.585, 4.677, 5.494, 4.665, 5.390, 5.294, 4.490, 4.796, 5.219,
           4.763, 5.074, 4.949, 5.514, 4.891, 4.571, 5.269, 5.210, 5.165, 5.403, 5.042, 5.081, 5.143, 5.642, 5.689]

mean_volt = mean(voltage)

print(f'Mean Voltage: {mean_volt:.3f}')
```

```python
stdev_volt = stdev(voltage)

print(f'Standard Deviation of Voltage: {round(stdev_volt,2):.3f}')
```

```python
mean_err = stdev_volt/(len(voltage)**(1/2))
print(f'Statistical Error in The Mean Voltage: {mean_err:.3f}')
```

```python
length = [0.400, 0.600, 0.800, 1.000, 1.200, 1.400, 1.600]

freq = [391.77, 261.73, 197.91, 157.17, 129.21, 110.97, 102.13]

plt.figure(1)
plt.plot(length, freq, 'o', color= 'red', markersize=4)
plt.xlabel('Length(m)', fontsize=14)
plt.ylabel('Frequency(Hz)', fontsize=14)
plt.show()
```

```python
def function_fit(length, a):
    frequency = a/length
    return frequency 

popt, pcov = curve_fit(function_fit, length, freq)

a = popt[0]

print('The best fit parameters are:\n')
print(f'a = {a}')
```

```python
xvals = np.linspace(0.4, 1.6, 200)
plt.figure(2)
plt.plot(length, freq, 'o', color='red', markersize=4, label='data')
plt.xlabel('Length(m)', fontsize=14)
plt.ylabel('Frequency(Hz)', fontsize=14)
plt.plot(xvals, function_fit(xvals, a), color='black', linewidth=1.2, label='best-fit model')
plt.legend()
plt.show()




```

```python
perr = np.sqrt(np.diag(pcov))
delta_a = perr[0]
print(f"The error of parameter a is {delta_a}")
```

```python
plt.figure(1)
plt.plot(length,freq-function_fit(length, a), 'o', color= 'red', markersize=4)
plt.title('Residuals')
plt.show()
```
